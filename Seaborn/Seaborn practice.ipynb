```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head(3)
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)  # remove kde plot , dont increase bins too much 
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex') #kind= kde null reg
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')  # joint plot for all column combinations
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])  # getting only kde plot
```

```python
tips.head(3)
```

```python
import numpy as np
```

```python
sns.barplot(x=tips['sex'],y=tips['total_bill'],data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
 sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
 
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```
